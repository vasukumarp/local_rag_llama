```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
import os
```

```python
from google.colab import userdata

groq_api_key = userdata.get('groq_api_key')
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    api_key= groq_api_key,
    model="llama-3.1-70b-versatile",
    temperature=0.2,
)
```

```python
loader = PyPDFLoader("/content/lora_documentation.pdf")

data = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=250,
    length_function=len,)

documents = text_splitter.split_documents(data)
```

```python
embedding = HuggingFaceEmbeddings()
```

```python
persist_directory = 'chroma_db'

vector_db = Chroma.from_documents(
    documents=documents,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
vector_db = Chroma(persist_directory=persist_directory, embedding_function=embedding)
```

```python
retriever = vector_db.as_retriever(search_kwargs={"k": 2})
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
from langchain.prompts import PromptTemplate
```

```python
def retrieval(question):

  system_prompt = """
  You are an intelligent and reliable AI assistant tasked with helping users efficiently retrieve accurate information from a vector database.
  Your goal is to understand user input deeply, interpret the context provided, and use it to retrieve the most relevant and helpful information.

  Instructions:
  - Carefully assess the context provided to grasp the topic or specific area of focus.
  - Use semantic search to identify the most relevant entries in the vector database, focusing on both direct matches and conceptually related information.
  - Ensure that responses are clear, concise, and directly address the user’s needs, minimizing extraneous information.

  Context: {context}
  """

  # Define the chat prompt template using 'input' as the input variable
  prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
  )

  # Create the question-answer chain
  question_answer_chain = create_stuff_documents_chain(llm, prompt)

  # Create the retrieval chain with a retriever
  chain = create_retrieval_chain(retriever, question_answer_chain)


  # Invoke the chain with the input query
  result = chain.invoke({"input": question})

  # Clean the output (remove newlines and excessive spaces)
  clean_output = result['answer'].replace("\n", " ").replace("  ", " ").strip()

  return clean_output
```

```python
question = "what is lora"

result = retrieval(question)
print(result)
```

```python
import gradio as gr


iface = gr.Interface(
    fn=retrieval,
    inputs="text",
    outputs="text",
    title="Q & A with Research Papers",
    description="Enter a query to retrieve relevant information from the vector database."
)

# Launch the interface
iface.launch()
```
